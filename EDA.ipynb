```python
import numpy as np 
import pandas as pd 

import random as rn

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
%matplotlib inline
```

```python
X_test = pd.read_csv("/home/birnesh/Documents/MNIST/inputs/test.csv")
train = pd.read_csv("/home/birnesh/Documents/MNIST/inputs/train.csv")
```

```python
y_train = train["label"]
X_train = train.drop(["label"],axis = 1)
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
print("Information about X_train :")
print(X_train.info())
print("*"*50)
print("Information about X_test:")
print(X_test.info())
print("*"*50)
print("Shape of y_train:")
print(y_train.shape)
```

```python
fig, axs = plt.subplots(1, 5, sharex=True, sharey=True, figsize=(10,6))
axs = axs.flatten()
for i in range(0,5):
    im = X_train.iloc[i]
    im = im.values.reshape(-1,28,28,1)
    axs[i].imshow(im[0,:,:,0], cmap=plt.get_cmap('gray'))
    axs[i].set_title(y_train[i])
plt.tight_layout()  
```

```python
li_idxs = []
for i in range(10):
    for nr in range(10):
        ix = y_train[y_train==nr].index[i]
        li_idxs.append(ix) 
```

```python
fig, axs = plt.subplots(10, 10, sharex=True, sharey=True, figsize=(10,12))
axs = axs.flatten()
for n, i in enumerate(li_idxs):
    im = X_train.iloc[i]
    im = im.values.reshape(-1,28,28,1)
    axs[n].imshow(im[0,:,:,0], cmap=plt.get_cmap('gray'))
    axs[n].set_title(y_train[i])
plt.tight_layout()    
```
